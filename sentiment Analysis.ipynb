```python
import nltk
import random
from nltk.corpus import movie_reviews
```

```python
documents= []
for cat in movie_reviews.categories():
    for fileid in movie_reviews.fileids():
        documents.append((list(movie_reviews.words(fileid)),cat))
```

```python
documents[3999]
```

```python
random.shuffle(documents)
```

```python
all_words =[]
for w in movie_reviews.words():
    all_words.append(w.lower())
```

```python
all_words = nltk.FreqDist(all_words)
```

```python
all_words.most_common(30)
```

```python
all_words["shit"]
```

```python
word_features = list(all_words.keys())[:2500]
```

```python
def find_features(document):
    words = set(document)
    features = {}
    for w in word_features:
        features[w] = (w in words)
    return features
```

```python
find_features(movie_reviews.words('neg/cv000_29416.txt'))
```

```python
featuresets = [(find_features(rev),category) for (rev,category) in documents]
```

```python
training_set = featuresets[:1900]
testing_set = featuresets[:1900]
```

```python
classifier =nltk.NaiveBayesClassifier.train(training_set)
```

```python
nltk.classify.accuracy(classifier, training_set)*100
```

```python
classifier.most_informative_features(20)
```

```python
import pickle
save_classifier = open("naivebayes.pickle","wb")
pickle.dump(classifier, save_classifier)
save_classifier.close()
```
